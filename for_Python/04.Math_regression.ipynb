## <strong>1 Sin() 함수의 회귀식 분석
데이터 생성, 2차 5차 sin 곡선을 활용한 모델의 예측

### <strong>01 데이터 생성하기
Sin() 곡선 그리기

```python
# 1. 데이터 생성
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Sin() 함수의 -2pi ~ 2pi 구간을 관찰
def f(x):                       
    return np.sin(x) + 0.5 * x

x = np.linspace(-2 * np.pi, 2 * np.pi, 50)  # (시작점, 종료점, Bins 구간 분할갯수)
plt.plot(x, f(x), 'b')                      # color : Blue
plt.grid(True); plt.xlabel('x'); plt.ylabel('f(x)')
```

### <strong>02. 1차 회귀식 예측
회귀식 모델의 다양한 적용 1

```python
# deg = 1 : 단항 기저함수 (Monomials as Basis Functions)
reg = np.polyfit(x, f(x), deg=1)  # np.polyfit() : 수식을 찾는다 [절편, 기울기1, ...] 
ry  = np.polyval(reg, x)          # np.polyval() : 수식계수를 데이터에 적용 (연산결과를 출력한다)
print('reg : {}\nry  : {}'.format(reg, ry))

plt.plot(x,  f(x),  'b',  label='f(x)')        # sin() 그래프
plt.plot(x,  ry,   'r.',  label='regression')  # np.polyfit() 회귀식 그래프
plt.legend(loc=0); plt.grid(True); plt.xlabel('x'); plt.ylabel('f(x)')
```

### <strong>03. 5차 다항식을 활용한 회귀식
회귀식 모델의 다양한 적용 2

```python
reg = np.polyfit(x, f(x), deg=5)  # [절편, 기울기1, 2, 3, 4, 5]
ry = np.polyval(reg, x)           # 회귀식 검증(추출 회귀식에 x를 대입한, 예측 Y값 출력)
print('reg : {}\nry  : {}'.format(reg, ry))

plt.plot(x, f(x),  'b',  label='f(x)')
plt.plot(x, ry,   'r.',  label='regression')
plt.legend(loc=0); plt.grid(True); plt.xlabel('x'); plt.ylabel('f(x)')
```

### <strong>04. 7차 다항식을 적용
회귀식 모델의 다양한 적용 3

```python
reg = np.polyfit(x, f(x), deg = 7)
ry = np.polyval(reg, x)
print('reg : {}\nry  : {}'.format(reg, ry))

plt.plot(x, f(x), 'b', label='f(x)')
plt.plot(x, ry,  'r.', label='regression')
plt.legend(loc=0); plt.grid(True); plt.xlabel('x'); plt.ylabel('f(x)')

print('\n일치확인',       np.allclose(f(x), ry))              # 7차 다항식을 정의한 결과 유사하지만, 완전히 일치하진 않는다
print('MSE(평균제곱오차)', np.sum((f(x) - ry) ** 2) / len(x))  # MSE(평균제곱오차) 로 Cost 계산 - 그 오차는 작다
```

## <strong>2 개별 기저함수
Individual Basis Functions : 위에선 1차, 3차, 5차 식을 <strong>기저함수</strong>로 활용하였다

```python
# http://darkpgmr.tistory.com/56
# 최소자승법(Least Square Method)은 모델, 데이터, residual2의 
# 합 또는 평균을 최소화하도록 파라미터를 결정하는 방법.

# Residual은 어떤 데이터가 추정된 모델로부터 얼마나 떨어진 값인가를 
# 나타내는 용어로 통계학에서는 '잔차'라고 부른다고 한다

# 최소자승법을 계산하는 방법은 크게 
# 해석학적 (analytic : 편미분 함수를 연립방정식으로 계산) 방법과 
# 대수적  (algebraic : 선형대수로 계산) 방법이 있다
# 모델    : 가설함수
# 파라미터 : 모델의 weight, bias 조절 값
```

```python
# Matrix를 기저함수로 활용
matrix = np.zeros((3 + 1, len(x)))
matrix[3, :] = x ** 3
matrix[2, :] = x ** 2
matrix[1, :] = x
matrix[0, :] = 1
print(matrix.shape)
```

### <strong> 01 최소 자승 최적화 문제를 풀이
<strong>고차원 다항식( 4 X 50 )</strong>을 활용 (잔차를 최소로 하는 회귀식을 계산)

np.linalg.lstsq(matrix.T, f(x))[0]

```python
reg = np.linalg.lstsq(matrix.T, f(x))[0]      # lstsq (least square) : 최소 자승(Least Square)의 해결
ry  = np.dot(reg, matrix)                     # 최소자승법 결과에 matrix를 적용한 결과를 출력 (np.dot(벡터 내적의 합))
print('reg : {}\nry  : {}'.format(reg, ry.shape))

plt.plot(x, f(x),  'b',  label='f(x)')        # sin() 그래프
plt.plot(x,  ry,  'r.',  label='regression')  # np.dot() 회귀식 연산결과
plt.legend(loc=0); plt.grid(True); plt.xlabel('x'); plt.ylabel('f(x)')
```

### <strong>02 기저함수에 sin() 함수를 적용
matrix[3, :] = np.sin(x)

```python
matrix[3, :] = np.sin(x)
reg = np.linalg.lstsq(matrix.T, f(x))[0]  # matrix.T : 전치행렬
ry = np.dot(reg, matrix)
print('reg : {}\nry  : {}'.format(reg, ry.shape))

plt.plot(x, f(x), 'b', label='f(x)')
plt.plot(x, ry,  'r.', label='regression')
plt.legend(loc=0)
plt.grid(True); plt.xlabel('x'); plt.ylabel('f(x)')
print('\n일치확인',       np.allclose(f(x), ry))              # 완전히 일치한다
print('MSE(평균제곱오차)', np.sum((f(x) - ry) ** 2) / len(x))  # MSE(평균제곱오차) Cost 계산 : 오차는 존재
```

## <strong>3 잡음이 포함된 경우
Noisy Data

```python
xn = np.linspace(-2 * np.pi, 2 * np.pi, 50)             # nd.array
xn = xn + 0.15 * np.random.standard_normal(len(xn))     # 잡음을 포함한 데이터 
yn = f(xn) + 0.25 * np.random.standard_normal(len(xn))  # Sin() 곡선에 적용한 array
```

```python
reg = np.polyfit(xn, yn, 7)
ry = np.polyval(reg, xn)

plt.plot(xn, yn, 'b^', label='f(x)')
plt.plot(xn, ry, 'ro', label='regression')
plt.legend(loc=0)
plt.grid(True); plt.xlabel('x'); plt.ylabel('f(x)')
# 결론 : 회귀법을 통하면 일정한 잡음을 제거할 수 있다
```

Regression with unsorted data

<strong>Multiple Dimensions</strong> : <strong>'최소 자승 회귀법'</strong>의 장점은, <strong>다차원 자료</strong>에도 적용이 가능하다

### <strong>01 데이터 생성
<strong>np.meshgrid(x, y) : 1차원 배열을 정방(square) 행렬로 선언

```python
# 1. np.meshgrid(x, y) : 1차원 배열을 정방(square) 행렬로 선언
x = np.linspace(0, 10, 10)
y = np.linspace(0, 10, 10)

X, Y = np.meshgrid(x, y) # 1차원 배열에서, 2차원 그리드를 생성 (해당 배열을 정방(square) 행렬로 선언)
plt.subplots(1, 1, figsize=(2, 2))
plt.plot(X);  X[:2]  # X는 다차원 데이터 10개의 병행 데이터
plt.plot(Y);  Y[:3]  # Y 는 축척 기울기 (np.meshgrid()) 
```

### <strong> 02 모델의 적용 
Function with two parameters

```python
def fm(x, y):
    return np.sin(x) + 0.25 * x + np.sqrt(y) + 0.05 * y ** 2

# 1차원 배열에서, 2차원 그리드를 생성 (해당 배열을 정방(square) 행렬로 선언)
x = np.linspace(0, 10, 20)
y = np.linspace(0, 10, 20)
X, Y = np.meshgrid(x, y)
print('X :', X.shape," Y :", Y.shape)

Z = fm(X, Y)    # 다차원 함수를 적용
x = X.flatten() # 2차원 Matrix 를, 1차원 array로 이어 붙인다 (fully connected)
y = Y.flatten()
print('x :', x.shape," y :", y.shape )
```

### <strong>03 회귀식 결과를 시각화
Visualization  https://matplotlib.org/mpl_toolkits/mplot3d/tutorial.html

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl

fig = plt.figure(figsize=(9, 6))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z, rstride=2, cstride=2, cmap=mpl.cm.coolwarm,
        linewidth=0.5, antialiased=True)
ax.set_xlabel('x'); ax.set_ylabel('y'); ax.set_zlabel('f(x, y)')
fig.colorbar(surf, shrink=0.5, aspect=5)
```

### <strong>04 기저함수 matrix
예상되는 여러특성을 포함하는 <strong>기저함수 matrix</strong>를 정의

```python
matrix = np.zeros((len(x), 6 + 1))
matrix[:, 0] = 1           # 1 (상수)
matrix[:, 1] = x           # x 배열 (직선)
matrix[:, 2] = y           # y 배열
matrix[:, 3] = x ** 2      # x 배열의 제곱 (2차곡선)
matrix[:, 4] = y ** 2      # y 배열의 제곱
matrix[:, 5] = np.sin(x)   # x sin() (sin곡선)
matrix[:, 6] = np.sqrt(y)  # y 제곱근  (포물선)
print(matrix.shape); matrix[:4]
```

### <strong>05 sm.OLS()
최소 자승 회귀계산 (회귀식과 함께, 다양한 추가정보를 제공)

cf) 결정계수 (http://math7.tistory.com/123)
1. r-squared : R^2 결정계수
1. 회귀식에서 개별데이터의 MSE 오차 거리거리로 0 ~ 1 사이의 값을 갖는다
1. 0 : 회귀식의 정확도가 낮다
1. 1 : 회귀식의 정확도가 높다

<img src ="http://cfile3.uf.tistory.com/image/2421B33654D318480A6D0D" align='left' width='400'>

```python
import statsmodels.api as sm
model = sm.OLS(fm(x, y), matrix).fit()

print("""OLS() 결정계수 : {} 
       \nOLS() 분석결과 중 최적 회귀계수만 출력 :\n {}
       \nOLS() 분석결과 요약 : {}
         """.format(model.rsquared,
                    model.params,
                    model.summary()))
```

### <strong>06 sm.OLS() 분석 결과를 활용
최적 회귀계수와, 독립 변수 좌표를 활용한 회귀함수 계산 및 시각화

```python
def reg_func(a, p):
    x, y = p
    f0 = a[0] * 1
    f1 = a[1] * x
    f2 = a[2] * y
    f3 = a[3] * x ** 2
    f4 = a[4] * y ** 2
    f5 = a[5] * np.sin(x)
    f6 = a[6] * np.sqrt(y)
    return (f6 + f5 + f4 + f3 + f2 + f1 + f0)

a = model.params
RZ = reg_func(a, (X, Y))

fig = plt.figure(figsize=(9, 6))
ax = fig.gca(projection='3d')
surf1 = ax.plot_surface(X, Y, Z, rstride=2, cstride=2,
            cmap=mpl.cm.coolwarm, linewidth=0.5,
            antialiased=True)
surf2 = ax.plot_wireframe(X, Y, RZ, rstride=2, cstride=2,
                          label='regression')
ax.set_xlabel('x');       ax.set_ylabel('y')
ax.set_zlabel('f(x, y)'); ax.legend()
fig.colorbar(surf, shrink=0.5, aspect=5)
```
